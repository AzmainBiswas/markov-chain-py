```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RandomWalk(p, start, NumberOfStape):
    Position = [start]
    for _ in range(NumberOfStape):
        u = np.random.uniform()
        if u < p:
            position = Position[-1] + 1
        else:
            position = Position[-1] - 1
        Position.append(position)
    return Position
    
```

For $p<0.5$

```python
p = 0.45

fig = plt.figure(figsize=(10, 3))
plt.plot(RandomWalk(p,0, 1000), color="green")
plt.axhline(y=0, color="r", linestyle="--")
plt.xlabel("Time",fontsize=15)
plt.ylabel("State of Random Walk",fontsize=15)
plt.title(f"1D Simple Random Walk for p = {p}",fontsize=15)
plt.grid(True)
plt.show()
```

For $p>0.5$

```python
p = 0.56


fig = plt.figure(figsize=(10, 3))
plt.plot(RandomWalk(p,0, 1000), color="green")
plt.axhline(y=0, color="r", linestyle="--")
plt.xlabel("Time",fontsize=15)
plt.ylabel("State of Random Walk",fontsize=15)
plt.title(f"1D Simple Random Walk for p = {p}",fontsize=15)
plt.grid(True)
plt.show()
```
