```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GamberRuinedGame(starting_amount, winning_amount, number_of_game, probability_of_win):
    t = 0
    intermediate_amount = starting_amount
    amounts = [intermediate_amount]
    while number_of_game > 0:
        t += 1
        u = np.random.uniform(size=1)
        if u < probability_of_win:
            intermediate_amount += 1
        else:
            intermediate_amount -= 1
        amounts.append(intermediate_amount)
        if intermediate_amount == 0:
            break
        if intermediate_amount == winning_amount:
            break
        number_of_game = number_of_game - 1
    return amounts
```

```python
starting_amount = 50
winning_amount = 150
number_of_game = 10000
```

For $p>0.5$

```python
p = 0.52
gamber_ruined_game = GamberRuinedGame(starting_amount,winning_amount,number_of_game,p)

fig = plt.figure(figsize=(10, 4))
plt.plot(gamber_ruined_game, color="green")
plt.axhline(y=0, color="r", linestyle="--")
plt.axhline(y=winning_amount, color="r", linestyle="--")
plt.xlabel(f"Number of game",fontsize=12)
plt.ylabel(f"State of Gambeler)",fontsize=12)
plt.title(f"Gambler's Ruin Problem (p = {p})",fontsize=12)
plt.grid(True)
plt.show()

print(f" Number of game: {len(gamber_ruined_game)} \n Winning or Lossing amount: {gamber_ruined_game[-1]} ")
```
